```python
from transformers import AutoModelForQuestionAnswering, AutoTokenizer, pipeline
```

```python
from datasets import load_dataset
dataset = load_dataset("wikipedia", "20220301.de")
artikel_mit_suchwort = []
suchwort = input("Bitte gib ein Wort ein, nach dem im Artikel gesucht werden soll: ")
gesammelter_text =''
model_name = "deepset/roberta-base-squad2"


nlp = pipeline('question-answering', model=model_name, tokenizer=model_name)

notFound = True

while(notFound):

  # Durchlaufe alle Artikel und prüfe, ob das Wort im Text enthalten ist
  for article in dataset['train']:
      if suchwort.lower() in article['title'].lower():
          artikel_mit_suchwort.append({
              'title': article['title'],
              'text': article['text']
          })
          gesammelter_text += article['text']
          QA1_input = {
            'question': 'Wurde folgende frage beantwortet: Wie viele Einwohner hat Deutschland? Antworte nur mit "True" oder "False',
            'context': gesammelter_text
          }
          res1 = nlp(QA1_input)
          print(res1)

# Überprüfen, ob Artikel gefunden wurden
if artikel_mit_suchwort:
    print(f"Es wurden {len(artikel_mit_suchwort)} Artikel gefunden, die das Wort '{suchwort}' enthalten:")

else:
    print("Es wurden keine Artikel gefunden.")


##QA2_input = {
    #'question': 'Wie viele Einwohner hat Deutschland?',
    #'context': gesammelter_text
#}
res2 = nlp(QA2_input)
print(res2)

```
